## Análise e Resultados

Observações: -No decorrer do projeto, procurei por possíveis soluções para filtrar os NaNs. Tive a idéia de trocar os NaNs no meu dataframe por um outro valor (no caso value='-'), para que não fosse omitido as linhas que continham NaNs, no intuito de tentar preservar os dados o máximo possivel. Feito essa estratégia, percebi que quando eu trocava os valores de NaN por '-', dados que eram dtype float, como a renda, se transformavam em object, e paravam de ser processados como desejados, além de criar uma nova categoria '-' nos gráficos que eu plotava. Decidi então optar pela opção de remover os NaNs, porêm, quando apliquei o dropna para remover os NaNs da mihna análise, me deparei que quando eu removia as linhas com NaN, todas as pessoas que não tinham internet em casa também eram desconsideradas (no caso todas as linhas que tinham 'Não' nessa variavel também tinham NaNs na mesma linha, portanto todos os 'Não' também eram removidos), por isso, para não prejudicar a análise, optei por não remover os NaNs.

-Apesar de ter sucesso com os Subplots de alguns gráficos de pizza, quando fui fazer com os gráficos crosstab de barra, n consegui fazer funcionar.
